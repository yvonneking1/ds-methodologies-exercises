1. Attendance Data

    Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

    You should end up with something like this:


name

Billy    0.5250

Jane     0.6875

John     0.9125

Sally    0.7625

Name: grade, dtype: float64

```python
attendance = pd.read_csv("attendance.csv")
```

```python
attendance
```

```python
attendance_melt = attendance.melt(id_vars="Unnamed: 0", var_name="date", value_name="attendance")
```

```python
attendance_melt.attendance = attendance_melt.attendance.str.replace("P","1")
```

```python
attendance_melt.attendance = attendance_melt.attendance.str.replace("H","0.5")
```

```python
attendance_melt.attendance = attendance_melt.attendance.str.replace("A","-1")
```

```python
attendance_melt.attendance = attendance_melt.attendance.str.replace("T",".9")
```

```python
attendance_melt.attendance = attendance_melt.attendance.astype("float")
```

```python
attendance_melt = attendance_melt.rename(columns={"Unnamed: 0": "name"})
```

```python
attendance_melt
```

```python
attendance_melt.groupby("name").mean()
```

3. Cake Recipes

    a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.


```python
cakes = pd.read_csv("cake_recipes.csv")
```

```python
cakes
```

b. Tidy the data as necessary.

```python
recipe_pos = cakes["recipe:position"].str.split(":", expand=True)
```

```python
recipe_pos.columns = ["recipe", "postition"]
```

```python
recipe_pos
```

```python
cakes = pd.concat([recipe_pos, cakes], axis=1).drop(columns="recipe:position")
```

```python
cakes
```

```python
cake_melt = cakes.melt(id_vars=["recipe", "postition"], var_name="temp", value_name="score")
```

```python
cake_melt
```

c. Which recipe, on average, is the best? recipe b

```python
cake_melt.groupby("recipe").mean()
```

d. Which oven temperature, on average, produces the best results? 275

```python
cake_melt.groupby("temp").mean()
```

e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees
